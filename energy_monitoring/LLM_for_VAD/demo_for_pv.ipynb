```python
import base64
import requests


# Function to encode the image
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')

# your key
api_key = ''

image_path = "./dataset/Faulty_solar_panel/train/1-Bird-drop/Bird (13).jpg"
base64_image = encode_image(image_path)
headers = {
  "Content-Type": "application/json",
  "Authorization": f"Bearer {api_key}"
}

text = "Is the PV panel in this picture anomalous? There are five possible anomalies: 1. Bird-dropped; 2. Dusty; 3. Snow-covered; 4. Electrical-damaged; 5. Physical-damaged. Please give your answer and explain it."

payload = {
  "model": "gpt-4o-mini",
  "messages": [
    {
      "role": "user",
      "content": [
        {
          "type": "text",
          "text": text
        },
        {
          "type": "image_url",
          "image_url": {
            "url": f"data:image/jpg;base64,{base64_image}"
          }
        }
      ]
    }
  ],
}

response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)

output = response.json()
print(output['choices'][0]['message']['content'])
```

```python
text = '''Is the PV panel in this picture anomalous? There are five possible anomalies: 1. Bird-dropped; 2. Dusty; 3. Snow-covered; 4. Electrical-damaged; 5. Physical-damaged. 
Please give your answer. Just give me a number from 0 to 6, where 0 indicates no anomaly. 1 is Bird-dropped; 2 is Dusty; 3 is Snow-covered; 4 is Electrical-damaged; 5 is Physical-damaged. 
'''

import base64
import requests
import os


# Function to encode the image
def encode_image(image_path):
    with open(image_path, "rb") as image_file:
        return base64.b64encode(image_file.read()).decode('utf-8')

# your key
api_key = ''

def get_all_filenames(directory_path):
    # Get a list of all files in the directory
    filenames = os.listdir(directory_path)
    
    # Filter out directories (if you only want files)
    filenames = [f for f in filenames if os.path.isfile(os.path.join(directory_path, f))]
    
    return filenames

# performance validation
# directory_path = ['./dataset/Faulty_solar_panel/train/0/',
#                   './dataset/Faulty_solar_panel/train/1-Bird-drop/',
#                   './dataset/Faulty_solar_panel/train/2-Dusty/',
#                   './dataset/Faulty_solar_panel/train/3-Snow-Covered/',
#                   './dataset/Faulty_solar_panel/train/4-Electrical-damage/',
#                   './dataset/Faulty_solar_panel/train/5-Physical-Damage/']

# transferability validation
directory_path = ['./dataset/Faulty_solar_panel/transfer/0/',
                  './dataset/Faulty_solar_panel/transfer/1-Bird-drop/',
                  './dataset/Faulty_solar_panel/transfer/2-Dusty/',
                  './dataset/Faulty_solar_panel/transfer/3-Snow-Covered/',
                  './dataset/Faulty_solar_panel/transfer/4-Electrical-damage/',
                  './dataset/Faulty_solar_panel/transfer/5-Physical-Damage/']

num = 20
inputs = []
for i in range(len(directory_path)):
  filenames = get_all_filenames(directory_path[i])[:num]
  for j in range(len(filenames)):
    inputs.append(directory_path[i]+filenames[j])

pred = []
truth = [0]*num + [1]*num + [2]*num + [3]*num + [4]*num + [5]*num


for fn in inputs:
  base64_image = encode_image(fn)
  headers = {
    "Content-Type": "application/json",
    "Authorization": f"Bearer {api_key}"
  }

  payload = {
    "model": "gpt-4o-mini",
    "messages": [
      {
        "role": "user",
        "content": [
          {
            "type": "text",
            "text": text
          },
          {
            "type": "image_url",
            "image_url": {
              "url": f"data:image/jpg;base64,{base64_image}"
            }
          }
        ]
      }
    ],
  }

  response = requests.post("https://api.openai.com/v1/chat/completions", headers=headers, json=payload)

  output = response.json()
  pred.append(int(output['choices'][0]['message']['content']))
```
